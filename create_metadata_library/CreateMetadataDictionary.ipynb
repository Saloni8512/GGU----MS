```python
#Return true if our current value is NaN

def isNan(a):
    return a != a
```

```python
#generate dictionary 
#key is metadata field
#value is a list of term1, term2, term3
# we use this to match metadata fields to found terms in our sentence output document

def metadata_term_library(df_metadata):
    
    dic_metadata_term = {}
    dic_metadata_category = {}
    
    for i,j in df_metadata.iterrows():
        
        dic_metadata_category[j['Combined Metadata Categorization']] = [j['Category']]
        dic_metadata_term[j['Combined Metadata Categorization']] = [j['Term1']]
        
        if not isNan(j['Sub-Category']):
            dic_metadata_category[j['Combined Metadata Categorization']].append(j['Sub-Category'])
        else:
            pass
        
        
        if not isNan(j['Term2']):
            dic_metadata_term[j['Combined Metadata Categorization']].append(j['Term2'])
        else:
            pass
        
        if not isNan(j['Term3']):
            dic_metadata_term[j['Combined Metadata Categorization']].append(j['Term3'])
        else:
            pass
        
    return(dic_metadata_term, dic_metadata_category)
        
    
    
```

```python
x,y = metadata_term_library(metadata_file)
```
